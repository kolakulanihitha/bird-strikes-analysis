```python
df['Cost: Total $'] = pd.to_numeric(df['Cost: Total $'].str.replace(',', ''), errors='coerce')
df['Feet above ground'] = pd.to_numeric(df['Feet above ground'], errors='coerce')
```

```python
mean_all = df.mean(numeric_only=True)

print("Mean values for all numerical columns:")
print(mean_all)
```

```python
median_all = df.median(numeric_only=True)

print("Median values for all numerical columns:")
print(median_all)
```

```python
# Calculate the mode for the 'Aircraft: Make/Model' column
mode_value = df['Aircraft: Make/Model'].mode()

print(f"The mode of the 'Aircraft: Make/Model' column is: {mode_value.iloc[0]}")
```

```python
# Calculate the mode for the 'Wildlife: Species' column
mode_value = df['Wildlife: Species'].mode()

print(f"The mode of the 'Wildlife: Species' column is: {mode_value.iloc[0]}")
```

```python
# Calculate the mode for the 'Altitude bin' column
mode_value = df['Altitude bin'].mode()

print(f"The mode of the 'Altitude bin' column is: {mode_value.iloc[0]}")
```

```python
# Calculate the mode for the 'Pilot warned of birds or wildlife?' column
mode_value = df['Pilot warned of birds or wildlife?'].mode()

print(f"The mode of the 'Pilot warned of birds or wildlife?' column is: {mode_value.iloc[0]}")
```

```python
import matplotlib.pyplot as plt

value_counts = df['Pilot warned of birds or wildlife?'].value_counts()

# Plot the bar chart
plt.figure(figsize=(6, 4))
value_counts.plot(kind='bar', color='red')
plt.title('Frequency of Pilot Warnings About Birds or Wildlife')
plt.xlabel('Warning Status')
plt.ylabel('Count')
plt.tight_layout()
plt.legend()
plt.show()
```

```python

value_counts = df['Effect: Impact to flight'].value_counts()

# Plot the bar chart
plt.figure(figsize=(12, 8))
value_counts.plot(kind='bar', color='lightcoral')
plt.title('Frequency of Impact Effects on Flight')
plt.xlabel('Impact Effect')
plt.ylabel('Count')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
cost_by_effect = df.groupby('Effect: Impact to flight')['Cost: Total $'].sum()
plt.figure(figsize=(12, 8))
cost_by_effect.plot(kind='bar', color='dodgerblue')
plt.title('Total Cost by Impact Effect on Flight')
plt.xlabel('Impact Effect')
plt.ylabel('Total Cost ($)')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#histogram for the 'Cost: Total $' column
plt.figure(figsize=(10, 6))
plt.hist(df['Cost: Total $'].dropna(), bins=30, color='salmon', edgecolor='black')
plt.title('Histogram of Distribution of Total Costs')
plt.xlabel('Total Cost ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#  histogram for the 'Altitude bin' column
plt.figure(figsize=(12, 8))
df['Altitude bin'].value_counts().sort_index().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Histogram of Distribution of Altitude Bins')
plt.xlabel('Altitude Bin')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
#Distribution plot for the 'Cost: Total $' column
plt.figure(figsize=(12, 8))
sns.histplot(df['Cost: Total $'].dropna(), kde=True, color='purple', bins=30)
plt.title('Distribution of Total Costs')
plt.xlabel('Total Cost ($)')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
# Plot the distribution plot of 'Effect: Impact to flight'
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Effect: Impact to flight', palette='Pastel1', order=df['Effect: Impact to flight'].value_counts().index)
plt.title('Distribution of Impact Effects on Flight')
plt.xlabel('Impact Effect')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
skewness = df['Cost: Total $'].dropna().skew()

print(f"The skewness of the 'Cost: Total $' column is: {skewness}")
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Effect: Impact to flight_encoded'] = le.fit_transform(df['Effect: Impact to flight'].astype(str))

# Calculate the skewness of the encoded 'Effect: Impact to flight' column
skewness = df['Effect: Impact to flight_encoded'].skew()

print(f"The skewness of the 'Effect: Impact to flight' column is: {skewness}")
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df['Altitude bin'], df['Cost: Total $'], alpha=0.5)
plt.title('Scatter Plot of Altitude Bin vs. Total Cost')
plt.xlabel('Altitude Bin')
plt.ylabel('Total Cost ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
print(df[['Altitude bin', 'Cost: Total $']].head())

# Plot the box plot for 'Altitude bin' vs 'Cost: Total $'
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Altitude bin', y='Cost: Total $', palette='Set2')
plt.title('Box Plot of Total Costs by Altitude Bin')
plt.xlabel('Altitude Bin')
plt.ylabel('Total Cost ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
std_dev = df['Cost: Total $'].dropna().std()

print(f"The standard deviation of the 'Cost: Total $' column is: {std_dev}")
```

```python
sample_variance = df['Cost: Total $'].dropna().var()

print(f"The sample variance of the 'Cost: Total $' column is: {sample_variance}")
```

```python
population_variance = df['Cost: Total $'].dropna().var(ddof=0)

print(f"The population variance of the 'Cost: Total $' column is: {population_variance}")
```
